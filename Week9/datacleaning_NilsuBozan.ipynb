```python
stopwords_set = set(stopwords.words('english'))


def remove_stopwords(text):
    #Tokenizing the text into words
    words = word_tokenize(text)
    #Removing stopwords
    filtered_words = [word for word in words if word.lower() not in stopwords_set]
    #Joining the remaining words
    processed_text = ' '.join(filtered_words)
    return processed_text

df['content'] = df['content'].apply(remove_stopwords)

df.head()
```

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    #tokenizing the text to seperate words
    tokens = word_tokenize(text)
    lemmas = []
    for token, tag in nltk.pos_tag(tokens):
        pos = tag[0].lower() if tag[0].lower() in ['a', 'r', 'n', 'v'] else wordnet.NOUN #Map the POS tag to WordNet POS tag
        lemma = lemmatizer.lemmatize(token, pos=pos)
        lemmas.append(lemma)
    return ' '.join(lemmas) #Join the lemmas back into a single string

# Apply lemmatization to 'content' column in your DataFrame
df['content'] = df['content'].apply(lemmatize_text)

df.head()
```

```python
#performs part of speech tagging
def apply_pos_tagging(text):
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    return pos_tags


column_name = 'content' 
df[column_name + '_pos_tags'] = df[column_name].apply(apply_pos_tagging)

df.head()
```

```python
import spacy

#Prints token texts and it's part of speech tag
def print_pos(df, column_name):
    nlp = spacy.load('en_core_web_sm')
    text = df[column_name][0]
    doc = nlp(text)
    for token in doc:
        print(token.text, token.pos_)
print_pos(df, 'content')
```
